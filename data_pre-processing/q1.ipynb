```python
import pandas as pd
```

```python
constructors = pd.read_csv('../data/constructors.csv')
constructors.head()
```

```python
constructor_results = pd.read_csv('../data/constructor_results.csv')
constructor_results.tail(20)
```

```python
constructors_standings = pd.read_csv('../data/constructor_standings.csv')
constructors_standings.tail(20)
```

```python
races = pd.read_csv('../data/races.csv')
races.tail()
```

```python
# merge  constructor_results and constructors and races on constructorId and raceId
df1 = pd.merge(constructor_results, constructors, on='constructorId')
df2 = pd.merge(df1, races, on='raceId')
df2.head()
```

```python
final_df = df2.groupby(['year', 'name_x'])['points'].sum().reset_index()
final_df.tail(30)
```

```python
final_10_years = final_df[final_df['year'] >= 2010]
final_10_years.head()
```

```python
# plot the line graph for the above data frame, the constrictor should be shown as colour of the line
# put the legend on the left, and the x-axis should be year and y-axis should be points
# on the x-axis show all the years
# the title of the graph should be "Constructor points over the years"
# the x-axis label should be "Year"
# the y-axis label should be "Points"
# insert the colour pallet [#76cabb,#7d3fcf,#76d74a,#ce50bf,#d1d24a,#462879,#78d78f,#d14a7c,#5d883a,#7678d8,#dc6931,#79b0ce,#cd3f41,#556489,#b88c3c,#352334,#cbc997,#833e6b,#47593d,#c79dcf,#773b2b,#cd928c]

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
sns.set_palette(['#76cabb','#7d3fcf','#76d74a','#ce50bf','#d1d24a','#462879','#78d78f','#d14a7c','#5d883a','#7678d8','#dc6931','#79b0ce','#cd3f41','#556489','#b88c3c','#352334','#cbc997','#833e6b','#47593d','#c79dcf','#773b2b','#cd928c'])

plt.figure(figsize=(15, 8))
sns.lineplot(x='year', y='points', hue='name_x', data=final_10_years)
plt.legend(loc='upper left')
plt.title('Constructor points over the years')
plt.xlabel('Year')
plt.ylabel('Points')
plt.show()
```

```python
# pivot dataframe
df_pivot = final_10_years.pivot(index='year', columns='name_x', values='points')

# create stacked bar chart
ax = df_pivot.plot(kind='bar', stacked=True, figsize=(10, 6))

# set title and axis labels
ax.set_title('Formula One Teams Points by Year')
ax.set_xlabel('Year')
ax.set_ylabel('Points')

# set legend
ax.legend(title='Team')

# show chart
plt.show()
```
