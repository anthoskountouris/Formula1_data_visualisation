```python
import pandas as pd
```

```python
circuits = pd.read_csv('../data/circuits.csv')
circuits.head()
```

```python
races = pd.read_csv('../data/races.csv')
races.head()
```

```python
# merge the two dataframes on circuitId
merged_data = pd.merge(circuits, races, on='circuitId')
merged_data.head()
```

## Map

```python
# group by year
final_data = merged_data[['year','name_x','location','country', 'lat', 'lng']]
# grouped_data = grouped_data.groupby('year')
final_data.head()

```

```python
# rename the columns
final_data.columns = ['Year','Circuit','City','Country', 'Latitude', 'Longitude']
final_data.head()
```

```python
final_data['Year'] = final_data['Year'].astype(int)
```

```python
final_data[final_data['Year'] == 2023]
```

```python
print(final_data.columns)
```

```python
result_dict = {}

for year, group in final_data.groupby('Year'):
    races = []
    for index, row in group.iterrows():
        race = {
            'Circuit': row['Circuit'],
            'City': row['City'],
            'Country': row['Country'],
            'Coordinates': [row['Latitude'],row['Longitude']]
        }
        races.append(race)
    result_dict[year] = races

print(result_dict)
```

```python
import json

with open('circuits.json', 'w') as f:
    json.dump(result_dict, f, indent=4)
```
